```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://books.toscrape.com/',headers=headers).text
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
soup
```

```python
soup.find("h3").find("a").text
```

```python
soup.find("p")
```

```python
book_name=[]
for i in soup.find_all("h3"):
    book_name.append(i.find('a').text)
```

```python
book_name
```

```python
soup.find_all("p",class_="price_color")
```

```python
price=[]
for i in soup.find_all("p",class_="price_color"):
    price.append(i.text)
```

```python
price
```

```python
soup.find("p",class_="instock availability").text.strip()
```

```python
stock=[]
for i in soup.find_all("p",class_="instock availability"):
    stock.append(i.get_text(strip=True))
```

```python
stock
```

```python
df=pd.DataFrame({"book name":book_name,"Price":price,"Stock Availability":stock})
```
