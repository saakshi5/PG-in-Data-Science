```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
area=[200,330,100,500,700,900]
price=[12,10,11,15,18,20]
```

```python
plt.figure(figsize=(4,4))
plt.plot(area,price,marker="o")
plt.title("Area Price Plot")
plt.xlabel("Area in sq/m")
plt.ylabel("Price in lakhs")
plt.show()
```

```python
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y1,color="pink",label="sqrtline")
plt.plot(x,y2,color="red",label="log_line")
plt.legend()
```

```python
engine_mean=df['engine-size'].mean()
engine_median=df['engine-size'].median()
```

```python
plt.figure(figsize=(6,6))
plt.title('Cars Price')
df['engine-size'].hist(bins=10)
plt.axvline(engine_mean,color="yellow")
plt.axvline(engine_median,color="red")
plt.xlabel('engine-size')
plt.ylabel('frequency')

plt.grid(False)
```
