```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
sns.distplot(df['price'])
```

```python
pmn=df['price'].mean()
pmd=df['price'].median()
```

```python
plt.title('Price')
sns.distplot(df['price'],hist=False)
plt.axvline(pmn,label='mean',color='red')
plt.axvline(pmd,label='median',color='green')
plt.legend()
```
