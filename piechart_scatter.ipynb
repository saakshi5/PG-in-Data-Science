```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(7,7))
df['fuel-type'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(7,7))
df['body-style'].value_counts().plot(kind='barh')
```

```python
x=list(df['body-style'].unique())
x
```

```python
y=list(df['body-style'].value_counts())
y
```

```python
plt.bar(x,y)
plt.xticks(rotation=90)
```

```python
df['body-style'].value_counts().plot.pie(autopct="% 1.1f%%",explode=(0,0,0.2,0,0))
```

```python
df.corr()
```

```python
x=df['engine-size']
#x
y=df['price']
y
```

```python
#positive correlation
plt.figure(figsize=(6,6))
plt.title('Engine-size & Price')

plt.scatter(x,y)
```

```python
#negative correlation
plt.figure(figsize=(6,6))
plt.title('City mileage & Price')
plt.scatter(x=df['city-mpg'],y=df['price'])
plt.show()
```
